```python
# create some data 
x = np.linspace(0,10,100)
x[:10]
```

```python
# Use the same data to scatter plot 
fig, ax = plt.subplots()
ax.scatter(x,np.sin(x));
ax.scatter(x,np.cos(x));
ax.plot(x,np.sqrt(x));
ax.set(title= "Simple Plot",
      xlabel = 'x-axis',
      ylabel = 'y-axis');
```

```python
# Make a plot from dictionary
nut_butter_prices = {'Almond Butter' : 10,
                    'Peanut Butter' : 8,
                    "Cashew butter" : 12}
fig,ax =plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax.set(title= "Butter Prices",
      xlabel = 'Butter types',
      ylabel = 'Prices');
```

### Subplot option 1

```python
fig , ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                          figsize=(10,5))
# pot to each in different plot
ax1.plot(x,np.sin(x));
ax2.scatter(x,np.sin(x));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.sin(x));
```

### Which one should you use (pyplot vs matplotlib 00 method)
* When Plotting something quickly, okay to use the pyplot method
* When plotting something more advanced , use the 00 method

```python
over_50 =heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
# pyplot method 
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# 00 method mixed with pyplot
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100]);
```

```python
## 00 Method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot
ax.set(title="Heart Disease Plot",
      xlabel="Age",
      ylabel="Chol");
# Add a legend
ax.legend(*scatter.legend_elements(),title="Targert");

# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
# Adding a vertical line
ax.axvline(over_50["age"].mean(),
          linestyle="--");
```

```python
# Subplot of chol,age,thalach
fig ,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex= True)
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

ax0.set(title="heart Disease And cholestrol Levels",
       ylabel="cholestrol")

ax0.legend(*scatter.legend_elements(),title="Target")

ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")
ax0.axvline(x=over_50["age"].mean(),
          linestyle="--");


scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

ax1.set(title="heart Disease And Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")
ax1.axvline(x=over_50["age"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
fig.savefig("Heart Disease Analysis")
```

```python
# Subplot of chol,age,thalach
fig ,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize=(10,10),
                             sharex= True)
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

ax0.set(title="heart Disease And cholestrol Levels",
       ylabel="cholestrol")
# change the x axis limit
ax0.set_xlim([50,80])

ax0.legend(*scatter.legend_elements(),title="Target")

ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--")
ax0.axvline(x=over_50["age"].mean(),
          linestyle="--");


scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="summer")

ax1.set(title="heart Disease And Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate")

# change the x axis limit
ax1.set_xlim([50,80])

# change the y axis limit
ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements(),title="Target")

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")
ax1.axvline(x=over_50["age"].mean(),
          linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
fig.savefig("Heart Disease Analysis")
```
